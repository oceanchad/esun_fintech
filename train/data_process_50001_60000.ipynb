```python
import os
import collections

file_50001_60000 = {}
for filename in os.listdir('/content/train'):
    try:
        num = int(filename.split('_')[0])
    except:
        continue
    if num >= 50001 and num <= 60000:
        file_50001_60000[num] = filename
# file_50001_60000.sort()
print(file_50001_60000)
```

```python
with open ('/content/50001_60000_temp.txt', 'r') as file:
    lines = file.readlines()
mode = 0

label_change = [1,2]
label_isnull = [5,6,7,8,9,10,11,12,13]
label_nochange= [16,17,18,19,20,21,22]
label_nochange= []

for i in label_change:
    items = lines[i].split(' ')
    for i, item in enumerate(items):
        if i > 0:
            try:
                num, word = int(item[0:5]), item[6:7]
            except:
                continue
            # print(num, word)
            file_50001_60000[num] = f"{file_50001_60000[num][0:6]}{word}.{file_50001_60000[num].split('.')[1]}"
            # print(num, file_50001_60000[num])

for i in label_isnull:
    items = lines[i].split(' ')
    for i, item in enumerate(items):
        if i > 0:
            try:
                num = int(item[0:5])
            except:
                continue
            # print(num, word)
            file_50001_60000[num] = f"{file_50001_60000[num][0:6]}isnull.{file_50001_60000[num].split('.')[1]}"
            # print(num, file_50001_60000[num])

for i in label_nochange:
    items = lines[i].split(' ')
    for i, item in enumerate(items):
        if i > 0:
            try:
                num = int(item[0:5])
            except:
                continue
            # print(num, word)
            file_50001_60000[num] = f"{file_50001_60000[num][0:6]}isnull.{file_50001_60000[num].split('.')[1]}"
            # print(num, file_50001_60000[num])
```

```python
with open('/content/50001_60000.txt', 'w') as file:
    for k, v in sorted(file_50001_60000.items()):
        file.write(v)
        if v.find('60000') == -1:
            file.write('\n')
```
